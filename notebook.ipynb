## Exercise 2: plot using matplotlib

```python
x = d.iloc[:,0]
y = d.iloc[:,1]
fig = plt.scatter(x, y)
```

```python
# create new figure with one subplot
fig, ax = plt.subplots(1)
ax.scatter(x, y)
```

```python
# create category color mapping by species
unique_species = d['species'].unique()
d['color'] = d['species'].apply(lambda x: np.where(unique_species == x)[0][0])
# define markers for each category
markers = ['*', 's', 'o']
d['marker'] = d['species'].apply(lambda x: markers[np.where(unique_species == x)[0][0]])
```

```python
x1 = d.iloc[:,0]
y1 = d.iloc[:,1]
x2 = d.iloc[:,2]
y2 = d.iloc[:,3]
#
fig, axs = plt.subplots(1,2,
    figsize = (12,5), 
    sharex = 'all', sharey = 'all' # share axis
    )

# a
ax = axs[0]
ax.set_title('1)', loc = 'left')
ax.scatter(x1, y1, 
    c = d['color'], 
    marker = 'x',
    )
ax.set_xlabel(d.columns[0])
ax.set_ylabel(d.columns[1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# b
ax = axs[1]
ax.set_title('2)', loc = 'left')
ax.scatter(x2, y2, 
    c = d['color'], 
    marker = 'x',
    )
ax.set_xlabel(d.columns[2])
ax.set_ylabel(d.columns[3])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
