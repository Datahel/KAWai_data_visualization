# Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data:

```python
d = sns.load_dataset('iris') # dataset
d.head()
```

```python
ss = d.describe() # summary statistics
ss
```

## Exercise 1: plot using pandas

```python
d.plot(x = 'sepal_length', y = 'sepal_width', kind = 'scatter')
```

```python
# your code here
ss.plot(kind = 'bar')
```

## Exercise 2: plot using matplotlib

```python
x = d.iloc[:,0]
y = d.iloc[:,1]
fig = plt.scatter(x, y)
```

```python
# create new figure with one subplot
fig, ax = plt.subplots(1)
ax.scatter(x, y)
```

```python
#
fig, axs = plt.subplots(1,2, figsize = (12,5))

# a
ax = axs[0]
ax.set_title('1)', loc = 'left')
ax.scatter(x, y)
# b
ax = axs[1]
ax.set_title('2)', loc = 'left')
```
